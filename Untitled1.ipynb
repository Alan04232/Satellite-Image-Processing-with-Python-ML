```python
# Colab setup
!pip install astroquery astropy matplotlib

from astroquery.mast import Observations
from astropy.coordinates import SkyCoord
import astropy.units as u
from astropy.io import fits
from astropy.wcs import WCS
import matplotlib.pyplot as plt
import numpy as np

# -------------------------------
# Change this to any target (e.g. "NGC 1300", "M51", "NGC 6543")
target = "NGC 1300"
# -------------------------------

# Resolve target
coord = SkyCoord.from_name(target)
print(f"Resolved {target}: RA={coord.ra.deg:.4f}, Dec={coord.dec.deg:.4f}")

# Query JWST observations around the target
obs = Observations.query_region(coord, radius=0.02*u.deg)
obs = obs[obs['obs_collection'] == 'JWST']

if len(obs) == 0:
    raise ValueError("No JWST observations found for this target.")

print(f"Found {len(obs)} JWST observations")

# Get products for first observation
products = Observations.get_product_list(obs[0])
# Prefer science-calibrated FITS
science = Observations.filter_products(products,
                                       productType="SCIENCE",
                                       mrp_only=False)

manifest = Observations.download_products(science[:1], mrp_only=False)
fits_path = manifest[0]['Local Path']
print("Downloaded:", fits_path)

# --- Plot image ---
with fits.open(fits_path) as hdul:
    h = None
    for hh in hdul:
        if hh.data is not None and hh.data.ndim >= 2:
            h = hh
            break
    if h is None:
        raise RuntimeError("No image data found.")
    data = h.data
    if data.ndim == 3:  # if cube, take first slice
        data = data[0]
    wcs = WCS(h.header)

plt.figure(figsize=(7,7))
plt.subplot(projection=wcs)
plt.imshow(data, origin="lower", cmap="grey", vmin=np.percentile(data,1), vmax=np.percentile(data,99))
plt.xlabel("RA")
plt.ylabel("Dec")
plt.title(f"{target} (JWST)")
plt.colorbar(label="Flux")
plt.show()

# --- Try spectrum plot ---
found_spectrum = False
with fits.open(fits_path) as hdul:
    for h in hdul:
        if h.is_image or h.data is None:
            continue
        if isinstance(h, fits.BinTableHDU):
            cols = [c.upper() for c in h.columns.names]
            if "WAVELENGTH" in cols and "FLUX" in cols:
                wave = h.data["WAVELENGTH"]
                flux = h.data["FLUX"]
                plt.figure(figsize=(8,4))
                plt.plot(wave, flux, lw=0.7)
                plt.xlabel("Wavelength")
                plt.ylabel("Flux")
                plt.title(f"{target} Spectrum (JWST)")
                plt.show()
                found_spectrum = True
                break

if not found_spectrum:
    print("No simple 1D spectrum in this dataset.")
```

```python
#!pip install astroquery astropy matplotlib

from astroquery.mast import Observations
from astroquery.skyview import SkyView
import matplotlib.pyplot as plt
from astropy.io import fits
from astropy.visualization import make_lupton_rgb

#  Object to search (change this to any galaxy/nebula)
target = "NGC 1300"

# ------------------------
# 1) Get a pretty RGB image from SkyView (DSS or HST color survey)
# ------------------------
images = SkyView.get_images(position=target, survey=['DSS2 Red', 'DSS2 Blue', 'DSS2 IR'])

# Combine into RGB
r = images[0][0].data
g = images[1][0].data
b = images[2][0].data

rgb_image = make_lupton_rgb(r, g, b, Q=10, stretch=0.5)

plt.figure(figsize=(8,8))
plt.imshow(rgb_image, origin='lower')
plt.title(f"RGB Image of {target}")
plt.axis('off')
plt.show()

# ------------------------
# 2) Search MAST for spectra of this target
# ------------------------
obs_table = Observations.query_object(target, radius="0.02 deg")
specs = Observations.filter_products(obs_table, productSubGroupDescription="SPECTRA")

if len(specs) > 0:
    spec = specs[0]  # take the first spectrum
    manifest = Observations.download_products(spec['obsid'], productSubGroupDescription="SPECTRA")

    # Load spectrum
    spec_file = manifest['Local Path'][0]
    hdul = fits.open(spec_file)
    hdul.info()

    data = hdul[1].data
    wavelength = data['WAVELENGTH']
    flux = data['FLUX']

    plt.figure(figsize=(10,5))
    plt.plot(wavelength, flux, color='purple')
    plt.xlabel("Wavelength")
    plt.ylabel("Flux")
    plt.title(f"Spectrum of {target}")
    plt.show()
else:
    print(f"No spectra found in MAST for {target}")
```
